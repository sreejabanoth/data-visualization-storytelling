```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_excel("/content/Superstore_Dummy_Data.xlsx")

# Set style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 1. Sales vs Profit by Sub-Category
sns.barplot(data=df, x="Sales", y="Sub-Category", hue="Profit", dodge=False, palette="coolwarm")
plt.title("Sales vs Profit by Sub-Category")
plt.tight_layout()
plt.savefig("/content/Sales_vs_Profit_SubCategory.png")
plt.clf()

# 2. Profit by Region
sns.barplot(data=df, x="Region", y="Profit", palette="RdYlGn")
plt.title("Profit by Region")
plt.tight_layout()
plt.savefig("/content/Profit_by_Region.png")
plt.clf()

# 3. Monthly Sales Trend
df_sorted = df.sort_values("Order Date")
sns.lineplot(data=df_sorted, x="Order Date", y="Sales", marker='o')
plt.title("Monthly Sales Trend")
plt.tight_layout()
plt.savefig("/content/Sales_Trend_Over_Time.png")
plt.clf()

# 4. Discount vs Profit
sns.scatterplot(data=df, x="Discount", y="Profit", hue="Sub-Category", style="Region")
plt.title("Discount vs Profit")
plt.tight_layout()
plt.savefig("/content/Discount_vs_Profit.png")
plt.clf()

# 5. Top 5 Unprofitable Products
top_losses = df.sort_values("Profit").head(5)
sns.barplot(data=top_losses, x="Profit", y="Sub-Category", palette="Reds")
plt.title("Top 5 Unprofitable Products")
plt.tight_layout()
plt.savefig("/content/Top_Unprofitable_Products.png")
```
